```python
# import libraries
import requests
from bs4 import BeautifulSoup
```

```python
# url to scrape
url = "https://www.loctechng.com"
```

```python
# requesting page
result = requests.get(url)
```

```python
# HTML String
html_string = ""

# Check if request was successful
if result.status_code == 200:
    print("Request was successful")
    html_string = result.text
    
print(html_string)
```

```python
# use beautiful soup

ps = BeautifulSoup(html_string)
print(ps)
```

```python
# use name parameter to select by tag name
body = ps.find(name="body")
print(body)
```

```python
# Get all attributes with courses info
addr =[]
links = body.findAll(name="a")[38:63]
for link in links:
    if link.has_attr('href'):
        if "http" in link.attrs['href']:
            addr.append(link.attrs['href'])
            print (link.attrs['href'])
print (addr)
print(len(addr))
```

```python
addrs = list(set(addr))
print(len(addrs))
print(addrs)
```

```python
i = 1
for url in addrs:
    print("%d- url: %s " % (i, url))
    i += 1
    result = requests.get(url)
    html_string = ""
    # Check if request was successful
    if result.status_code == 200:
        print("Request was successful")
        html_string = result.text
        
        pss = BeautifulSoup(html_string)
        bodyt = pss.find(name="body")
        
        p = bodyt.findAll(name='p')
        strong = bodyt.findAll(name='strong')
        #print(p)

        if strong and len(strong) > 1:
            p = strong
        else:
            p = p

        #print(p)
        for par in p:
            if "Pre" in par.contents[0]:
                #print(par)
                next_ul_tag = par.findNext('ul').text
                next_ul_tag1 = par.findNext('ol').text
                if next_ul_tag or next_ul_tag1:
                    print(next_ul_tag)
            if "Post" in par.contents[0]:
                #print(par)
                next_ul_tag2 = par.findNext('ul')
                next_ul_tag3 = par.findNext('ol')
                if next_ul_tag or next_ul_tag3:
                    print(next_ul_tag2)
```
