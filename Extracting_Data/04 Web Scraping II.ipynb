```python
# import libraries
import requests
from bs4 import BeautifulSoup
```

```python
# url to scrape
url = "http://evaezekwem.com/"
```

```python
# requesting page
result = requests.get(url)
```

```python
# HTML String
html_string = ""

# Check if request was successful
if result.status_code == 200:
    print("Request was successful")
    html_string = result.text
    
print(html_string)
```

```python
# use beautiful soup

ps = BeautifulSoup(html_string)
```

```python
print(ps)
```

```python
# use name parameter to select by tag name
body = ps.find(name="body")
```

```python
print(body)
```

```python
# use text attribute to get the content of the tag
print(body.find(name="h1").text)
```

```python
# get all previous companies he worked
print(body.findAll(name="h4"))
```

```python
companies = body.findAll(name="h4")[:4]

print(companies)
```

```python
# Print all previous companies where you have worked
for comp in companies:
    print(comp.text)
```

```python
# get the various years he spend at these companies
years = body.findAll(name="h6")
print(years)
```

```python
# Print duration spent in previous companies where you have worked

duration = []
for year in years:
    #print(year.text)
    first = int(year.text.partition('-')[0])
    
    #print(first)
    try:
        second = int(year.text.partition('-')[2])
    
    except:
        second = first
        
    #print("Start: %d, End: %d"%(first, second))
    time_spent = second - first
    
    if time_spent == 0:
        duration.append(1)
    else:
        duration.append(time_spent)
        
print(duration)
```

```python
# Print companies and duration worked

for index, comp in enumerate(companies):
    print("Company: %s \nNo of Years: %d" % (comp.text, duration[index]))
```
