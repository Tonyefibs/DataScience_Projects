```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os
```

```python
# Stanford 2017 data
url = "https://www.collegedata.com/cs/data/college/college_pg02_tmpl.jhtml?schoolId=781"
```

```python
# requesting page
result = requests.get(url)
```

```python
# HTML String
html_string = ""

# Check if request was successful
if result.status_code == 200:
    print("Request was successful")
    html_string = result.text
```

```python
ps = BeautifulSoup(html_string)
print(ps)
```

```python
# use name parameter to select by tag name
body = ps.find(name="body")
```

```python
div1 = body.find(id='section8')
print(div1)
```

```python
tables = div1.findAll('table')
print(tables)
```

```python
html_str = []
for table in tables:
    html_str.append(table)
print(html_str[1])
```

```python
dfs = pd.read_html(html_str[0], header=0)
dfs[0]
```
